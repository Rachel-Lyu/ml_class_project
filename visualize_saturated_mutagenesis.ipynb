```python
### visualize mutagenesis
import numpy as np
import matplotlib.pyplot as plt

iterations = 15
num_seqs = 200
```

```python
all_fitness_vals = np.array(np.load('../results/sat_mut_200seq_15iter_all_fitnesses.npy', allow_pickle = True),dtype=float)
best_fitness_vals = np.array(np.load('../results/sat_mut_200seq_15iter_best_fitnesses.npy', allow_pickle = True), dtype=float)
val_fitness_vals = np.array(np.load('../results/sat_mut_200seq_15iter_val_fitnesses.npy', allow_pickle = True), dtype=float)
seqs = np.array(np.load('../results/sat_mut_200seq_15iter_seqs.npy', allow_pickle = True), dtype=float)
orig_seqs = np.array(np.load('../results/sat_mut_200seq_15iter_orig_seqs.npy', allow_pickle = True), dtype=float)
```

```python
import seq_functions


### make sequences as strings for Rachel

# with open('../results/satmut_orig_seqs_as_strings.txt', 'a') as the_file:
#     for ii in range(orig_seqs.shape[0]):
#         seq_as_string = seq_functions.inverse_onehot_seq(orig_seqs[ii,:,:])
#         the_file.write(''.join(seq_as_string) + '\n')
        
with open('../results/satmut_seqs_as_strings.txt', 'a') as the_file:
    for ii in range(seqs.shape[0]):
        seq_as_string = seq_functions.inverse_onehot_seq(seqs[ii,:,:])
        the_file.write(''.join(seq_as_string) + '\n')
#for ii in range(seqs.shape[0]):
#    seq_as_string = seq_functions.inverse_onehot_seq(seq)

```

```python
plt.figure(figsize=(10,6))
for ii in range(100):
    plt.plot(  range(iterations+1), best_fitness_vals[ii,:] )
    plt.xlabel('Iteration',fontsize = 18)
    plt.ylabel('Fitness',fontsize = 18)
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=16)
    
plt.savefig('../results/sat_mut_'+str(num_seqs)+'seqs_fitness_by_'+str(iterations)+'iter.png')
```

```python
plt.figure(figsize=(16,8))
for ii in range(100):
    plt.plot(  range(8), best_fitness_vals[ii,:8] )
    plt.xlabel('Iteration',fontsize = 16)
    plt.ylabel('Fitness',fontsize = 16)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
```

```python
plt.figure(figsize=(10,6))
for ii in range(100):
    plt.plot(  range(iterations+1), np.mean(val_fitness_vals[:,:,ii],axis = 1) )
    plt.xlabel('Iteration',fontsize = 18)
    plt.ylabel('Fitness',fontsize = 18)
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=16)
    
plt.savefig('../results/sat_mut_' + str(num_seqs) + 'seqs_val_fitness_by_' + str(iterations) + 'iter.png')
```
