```python
### visualize mutagenesis
import numpy as np
import matplotlib.pyplot as plt

iterations = 12
num_seqs = 100
```

```python
all_fitness_vals = np.array(np.load('../results/sat_mut_100seq_12iter_all_fitnesses.npy', allow_pickle = True),dtype=float)
best_fitness_vals = np.array(np.load('../results/sat_mut_100seq_12iter_best_fitnesses.npy', allow_pickle = True), dtype=float)
val_fitness_vals = np.array(np.load('../results/sat_mut_100seq_12iter_val_fitnesses.npy', allow_pickle = True), dtype=float)
seqs = np.array(np.load('../results/sat_mut_100seq_12iter_seqs.npy', allow_pickle = True), dtype=float)
```

```python
plt.figure(figsize=(16,8))
for ii in range(100):
    plt.plot(  range(iterations+1), best_fitness_vals[ii,:] )
    plt.xlabel('iteration',fontsize = 16)
    plt.ylabel('fitness',fontsize = 16)
    plt.xticks(fontsize=12)
```
